### AULA 02: ESTATÍSTICAS, VISUALIZAÇÃO DE DADOS E DISTRIBUIÇÕES

Renomeando a coluna droga por composto:

```python
mapa = {'droga': 'composto'}
dados.rename(columns=mapa, inplace=True)
```

```python
dados.head()
```

Identificando os cinco elementos mais frequentes:

```python
cod_compostos = dados['composto'].value_counts().index[0:5]
```

```python
cod_compostos
```

Filtrando a base de dados com a função query, informando a variável definida fora da função:

```python
dados.query('composto in @cod_compostos')
```

Importando a biblioteca Seaborn e plotando o gráfico:

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(8, 6))
ax = sns.countplot(x = 'composto', data=dados.query('composto in @cod_compostos'))
ax.set_title('Top 5 compostos')
plt.show()
```

**Desafio 02:** Melhorar a visualização alterando tamanho da fonte



*   Alterando a fonte e melhorando a visualização do gráfico countplot:



```python
sns.set()         
plt.rc('axes', titlesize=16)    
plt.rc('axes', labelsize=16)    
plt.rc('xtick', labelsize=12)    
plt.rc('ytick', labelsize=12)    
plt.figure(figsize=(8, 6))
ax = sns.countplot(x = 'composto', data=dados.query('composto in @cod_compostos'), order=cod_compostos)
ax.set_title('Top 5 compostos')
ax.set_ylabel('Contagem')
ax.set_xlabel('Composto')
plt.show()
```

**Desafio 04:** Estudar sobre as estatíticas retornadas no .describe()



> As análises estatísticas geradas pelo comando .describe() representam:

*   count - contagem dos valores existentes não nulos da amostragem
*   mean - média dos valores da amostragem
*   std - desvio padrão (uniformidade) da amostragem
*   min - valor mínimo da amostragem
*   25% - primeiro quartil
*   50% - mediana
*   75% - terceiro quartil
*   max - valor máximo da amostragem


**Desafio 05:** Refletir sobre a manipulação do tamanho das visualizações


> O tamanho das visualizações dos dados interfere diretamente na interpretação dos resultados. A legenda, a cor, a fonte, o espaçamento, entre outras manipulações podem alterar a interpretação dos dados feitas por quem irá receber aquelas visualizações. Estes dados distorcidos podem levar a erros e invalidar a análise. 

> Há fontes que utilizam visualizações de dados distorcidas com intuíto de chamar atenção de quem visualiza tais dados e até mesmo para induzí-los a uma análise que convenha e favorece uma determinada parte.
